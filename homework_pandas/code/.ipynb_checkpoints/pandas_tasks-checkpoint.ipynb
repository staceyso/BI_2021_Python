## Task 1

Построить диаграмму распределений нуклеотидов A, T, G и C из таблицы по ссылке.

```python
df = pd.read_csv("https://github.com/Serfentum/bf_course/raw/master/14.pandas/train.csv")
df
```

```python
plot = df[['pos', 'A', 'C', 'T', 'G']].plot.bar(x='pos',
                                               figsize=(15, 9),
                                               title="Nucleotide frequency for each position in read",
                                               xlabel="Position in read",
                                               ylabel="Nucleotide frequency",
                                               fontsize=10,
                                               stacked=True)
plot.axes.title.set_size(20)
fig = plot.get_figure()
fig.savefig("../data/task1_nucl_freq.png")
```

В датафрейме нет пропущенных значений.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sns.set_theme(context='notebook', style='darkgrid')
```

Построим гистограммы распределений количественных признаков:

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Numeric data distributions')

# age
sns.histplot(data['age'], ax=axes[0], kde=True, bins=25)
axes[0].set_title('Age')

# amount
sns.histplot(data['amount'], ax=axes[1], kde=True, bins=25)
axes[1].set_title('Amount')

# duration
sns.histplot(data['duration'], ax=axes[2], kde=True, bins=25)
axes[2].set_title('Duration');
```

Можем построить график зависимости срока кредита от возраста и личного статуса клиента:

```python
plt.figure(figsize=(15, 5))
sns.barplot(x='personal_status_sex', y='duration', data=data)
plt.title('Duration depending on personal status', size=18)
plt.xlabel('Sex and personal status', size=14)
plt.ylabel('Duration', size=14)
plt.show()
```

Также можем посмотреть на зависимость стажа от возраста:

```python
plt.figure(figsize=(15, 5))
sns.barplot(x='age', y='employment_duration', data=data)
plt.title('Employment duration depending on the age', size=18)
plt.xlabel('Age', size=14)
plt.ylabel('Employment duration', size=14)

plt.show()
```

Теперь посмотрим на корреляцию количественных признаков:

```python
NumericData = data._get_numeric_data()
```

```python
colormap = plt.cm.RdBu
plt.figure(figsize=(5,5))

plt.title('Pearson Correlation of Features', y=1.1, size=18)
sns.heatmap(NumericData.corr(),
            linewidths=0.1, vmax=1.0, 
            square=True, cmap=colormap, linecolor='white', annot=True)
plt.show()
```

А также можем проверить, как зависит  количества положительных и отрицательных кейсов в зависимости от размера сберегательного счёта:

```python
plt.figure(figsize=(15,7))
sns.histplot(x='savings', hue='credit_risk', palette=["green", "red"], multiple= "dodge", shrink=0.9, alpha=1, data=data)
plt.title('Credit risk depending on the amount of savings', size=18)
plt.xlabel('Savings', size=14)
plt.ylabel('Count', size=14);
```
