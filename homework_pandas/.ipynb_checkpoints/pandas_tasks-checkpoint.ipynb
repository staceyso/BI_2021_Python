```python
import pandas as pd
```

## Task 1

Построить диаграмму распределений нуклеотидов A, T, G и C из таблицы по ссылке.

```python
df = pd.read_csv("https://github.com/Serfentum/bf_course/raw/master/14.pandas/train.csv")
df
```

```python
plot = df[['pos', 'A', 'C', 'T', 'G']].plot.bar(x='pos',
                                               figsize=(15, 9),
                                               title="Nucleotide frequency for each position in read",
                                               xlabel="Position in read",
                                               ylabel="Nucleotide frequency",
                                               fontsize=10,
                                               stacked=True)
plot.axes.title.set_size(20)
fig = plot.get_figure()
fig.savefig("../data/task1_nucl_freq.png")
```
